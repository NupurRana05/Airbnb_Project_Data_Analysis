# visualization

# what is the distribution of listing prices ?

```python
plt.figure(figsize =(10,6))
sns.histplot(df['price'],bins=50,kde=True,color='red')
plt.title(" Distribution of listing price")
plt.xlabel("price$")
plt.ylabel("frequency")
plt.show()
```

```python
# The Histogram shows a fairly even distribution of listing prices across different price ranges, indicating no particular concentration of listings
# in any specific price range. The KDE line helps visualize this even spread more clearlly, confirming that the dataset contains listings with a wide
# variety of prices.


 
```

#  how are different room types distributed ?

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="room type", data=df, color="hotpink")
plt.title("room type distribution")
plt.xlabel("room type")
plt.ylabel("count")
plt.show()
```

   # what is the relationship between price and room type 

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.boxplot(x="room type", y="price", hue="room type", data=df, palette="Set1")

plt.title("Room Type vs Price")
plt.xlabel("Room Type")
plt.ylabel("Price ($)")
plt.legend(title="Room Type")   # Correct legend usage
plt.show()
```

# how has the number of reviews change over with time

```python
df.columns
```

```python
df.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert to datetime
df['last review'] = pd.to_datetime(df['last review'])

# Group reviews per month
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

# Plot
plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line', color='red')
plt.title("Number of Reviews Over Time")
plt.xlabel("Date (Monthly)")
plt.ylabel("Number of Reviews")
plt.show()
```
